```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import scipy.stats as st
import requests
import time
from scipy.stats import linregress
from pprint import pprint

file_to_load = "datasets_670134_1261707_data_by_year.csv"
```

```python
yearly_data = pd.read_csv(file_to_load)
```

```python
yearly_data.head()
```

```python
recent_years = yearly_data.loc[yearly_data["year"]>= 2010]
recent_years.head(11)
```

```python
year_group = recent_years.groupby(["year"])
year_group_df =pd.DataFrame(year_group)
year_group_df.info()
```

```python
# year = [2010,2011,2012,2013,2014,2015,2016,2017,2018,2019,2020]
```

```python
danceability, = plt.plot( recent_years["year"],recent_years["danceability"], label = "Danceability")
energy, =plt.plot( recent_years["year"],recent_years["energy"], label = "energy")
instrumentalness, = plt.plot(recent_years["year"],recent_years["instrumentalness"], label = "instrumentalness")
speechiness, = plt.plot(recent_years["year"],recent_years["speechiness"], label = "speechiness")
plt.title("Year vs. Different Music Attributes")
plt.xlabel("Year")
plt.ylabel("Music Attributes")
plt.legend(handles = [danceability, energy, instrumentalness, speechiness])
```

```python
popularity = plt.plot( recent_years["year"],recent_years["popularity"], label = "popularity")
plt.title("Year vs. Popularity of Songs")
plt.xlabel("Year")
plt.ylabel("Music Popularity")
```
